## 规则图
- 30 nodes
- degree of each node = 3

```python
import networkx as nx
import matplotlib.pyplot as plt
# regular graphy
# generate a regular graph which has 50 nodes & each node has 3 neghbour nodes.
RG = nx.random_graphs.random_regular_graph(3, 30)
# the spectral layout
pos = nx.spectral_layout(RG)
# draw the regular graphy
nx.draw(RG, pos, with_labels = False, node_size = 100)
plt.show()
```

## ER随机图
- 30 nodes
- connet a random pair with probility = 0.1


```python
import networkx as nx
import matplotlib.pyplot as plt
 
# erdos renyi graph
# generate a graph which has n=50 nodes, probablity p = 0.1.
ER = nx.random_graphs.erdos_renyi_graph(30, 0.1)
# the shell layout
pos = nx.shell_layout(ER)
nx.draw(ER, pos, with_labels = False, node_size = 100)
plt.show()
```

## 小世界
- 30 nodes
- each node has 4 neighbour nodes
- random reconnection probability was 0.3.


```python
import networkx as nx
import matplotlib.pyplot as plt

# WS network

# generate a WS network which has 50 nodes,
# each node has 4 neighbour nodes,
# random reconnection probability was 0.3.
WS = nx.random_graphs.watts_strogatz_graph(30, 4, 0.3)
# circular layout
pos = nx.circular_layout(WS)
nx.draw(WS, pos, with_labels = False, node_size = 100)
plt.show()
```

## 无标度
- 30 nodes
- add 1 edge each time

```python
import networkx as nx
import matplotlib.pyplot as plt

# BA scale-free degree network
# generalize BA network which has 30 nodes, m = 1
BA = nx.random_graphs.barabasi_albert_graph(30, 1)
# spring layout
pos = nx.spring_layout(BA)
nx.draw(BA, pos, with_labels = False, node_size = 30)
plt.show()
```
